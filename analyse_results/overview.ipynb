```python
deciles=list(set(survey['decile']))
deciles
```

```python
mean_wb_loss=[]
for dec in deciles:
    mult=survey.loc[survey['decile']==dec, 'wb_loss']*survey.loc[survey['decile']==dec, 'weight']
    mean_ls=mult.sum()/survey.loc[survey['decile']==dec, 'weight'].sum()
    mean_wb_loss.append(mean_ls)
mean_wb_loss
```

```python
import matplotlib.pyplot as plt

plt.scatter(deciles,np.array(mean_wb_loss)*100000)
```

```python
n_events=list(set(survey['n_events']))
n_events=n_events[:-4]
```

```python
results_dec_events=np.zeros((len(deciles), len(n_events)))
for i, dec in enumerate(deciles):
    for j, c_ev in enumerate(n_events):
        if c_ev == 12:
            mult=survey.loc[(survey['decile']==dec) & 
                            (survey['n_events']>=c_ev), 'wb_loss']*survey.loc[(survey['decile']==dec)&
                                                                              (survey['n_events']>=c_ev), 'weight']
            results_dec_events[i,j]=mult.sum()/survey.loc[(survey['decile']==dec) & (survey['n_events']>=c_ev), 'weight'].sum()

        else:
            mult=survey.loc[(survey['decile']==dec) & 
                            (survey['n_events']>=c_ev), 'wb_loss']*survey.loc[(survey['decile']==dec)&
                                                                              (survey['n_events']>=c_ev), 'weight']
            results_dec_events[i,j]=mult.sum()/survey.loc[(survey['decile']==dec) & (survey['n_events']>=c_ev), 'weight'].sum()
```

```python
for i,dec in enumerate(deciles):
    plt.plot(n_events,results_dec_events[i,:], label='dec: ' + str(dec))
plt.legend()
plt.ylabel('WB-Loss')
plt.xlabel('number events')
```
