# Housing

```python
house_data = pd.read_csv('/home/insauer/projects/WB_model/Data_Malawi/MWI_2010_IHS-III_v01_M_CSV/Full_Sample/Household/hh_mod_f.csv')
house_data = house_data[['case_id', 'hh_f04a', 'hh_f04b','hh_f18', 'hh_f25', 'hh_f26a', 'hh_f26b', 'hh_f35', 'hh_f37']]
house_data = house_data.replace(np.nan, 0)
house_data = house_data.groupby(['case_id']).sum()
house_data = house_data.rename(columns={'hh_f04a': 'rent', 'hh_f18': 'firewood', 'hh_f25': 'electricity',
                                        'hh_f35': 'cell_phones', 'hh_f37': 'drinking_water'})
house_data
```

```python
unit_dict = {3:365,
             4: 52,
             5: 12,
             6: 1,
             0: 0,
             1: 1}
def convert_unit(x):
    new_prize = x['rent']*unit_dict[x['hh_f04b']]
    
    return pd.Series(new_prize,
                     index=['housing'])
```

```python
unit_dict = {3:365,
             4: 52,
             5: 12,
             6: 1,
             0: 0,
             1: 1,
             np.nan: 0.0}
def convert_unit_e(x):
    if not np.isnan(x['hh_f26a']):
        print(x['electricity'])
        new_prize = (x['electricity']/x['hh_f26a']) *unit_dict[x['hh_f26b']]
    else:
        new_prize = 0
    
    return pd.Series(new_prize,
                     index=['electricity'])
```

```python
house_data['housing']=house_data.replace(np.nan,0).apply(convert_unit, axis=1)
house_data['electricity'] = house_data.apply(convert_unit_e, axis=1)
house_data['heat_electricity']=house_data['firewood']*52+house_data['electricity'].replace(np.nan,0)
house_data['telecommunication']=house_data['cell_phones']*12

house_data.loc[~house_data['electricity'].isna()]
```

```python
consum['housing']=house_data['housing']
consum['heat_electricity']=house_data['heat_electricity']
consum['drinking_water']=house_data['drinking_water']
consum['telecommunication']=house_data['telecommunication']
consum
```
